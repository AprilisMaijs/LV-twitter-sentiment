```python
#regexes - Alīna Govoruhina(2022)
newLine ="\\n|\\r"
urls = '(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})'
numbers = '\d+((\.|\-)\d+)?'
mentions = '\B\@([\w\-]+)'
hashtag = '#'
whitespaces = '\s+'
leadTrailWhitespace = '^\s+|\s+?$'

#combining regexes into single 
combined_pattern = f'{newLine}|{urls}|{numbers}|{mentions}|{hashtag}'
df['text'] = df['text'].str.replace(combined_pattern, '')
```

```python
# Split the dataset into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Initialize a tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
max_length = 25
tokenizer.padding = True
#tokenizer(max_length=max_length, truncation=True)

# Define a function to tokenize the text data

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=max_length)

# Convert the training and validation datasets to a format that can be used by transformers
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
print('got here')
train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# Print the first example in the training dataset
print(train_dataset[0])
```

#workin

```python
model = AutoModelForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=3)
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
from sklearn.metrics import f1_score, classification_report

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1_macro = f1_score(labels, preds, average='macro')
    class_report = classification_report(labels, preds, output_dict=True)
    return {'f1_macro': f1_macro, 'class_report': class_report}
```

```python
classification_report
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer_lvBERT", 
                                  evaluation_strategy="epoch",
                                  save_strategy='epoch',
                                  num_train_epochs=11
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer
)
```

```python
#search term for the relevant dates
year = '2016-08|2016-09|2016-10|2016-11|2016-12|2017-01|2017-02|2017-03|2017-04|2017-05|2017-06|2017-07'

#search terms for the political parties tracked that year
zzs = 'ZZS|zzs|zalie zemnieki|zaļie zemnieki|Zaļo un Zemnieku savienība|Zalie un zemnieki'
saskana = 'Saskaņa|Saskana|saskana|saskaņa'
vienotiba = 'Vienotība|Vienotiba'
nsl = 'NSL|no sirds Latvijai|No sirds latvijai|Sudraba'
kpv = 'KPV|Gobzems|kam pieder valsts|Kam pieder Valsts'
na = 'NA|nacionāļi|Nacionāļi|nacionali|Nacionālā apvienība|nacionala apvieniba'
lra = 'LRA|reģionu apvien|Latvijas Reģionu Apvienība|Bondars'

party_search_terms = [zzs,saskana,vienotiba,nsl,kpv,na,lra]

#new df with only time-relevant data and retweets removed
df_year = df[df['year_month_str'].str.contains(year)]
df_year = df_year.loc[~df_year['message'].str.startswith('RT')]

#new column in the df to mark the party that the tweet. If no party will be found named these fields will remain empty
df_year['party'] = ['' for i in range(df_year.shape[0])]
```

```python
import re
#atzīmē katram tvītam atbilstošās partijas nosaukumu
for i, row in df_year.iterrows():
    for term in party_search_terms:
        name = term.split('|')[0]
        #print(term)
        #print(name)
        if re.search(term, row['message']):
            print(i)
            df_year.at[i, 'party'] = name
```

```python
import torch

tokenizer = AutoTokenizer.from_pretrained("AiLab-IMCS-UL/lvbert")

i=0
sentiments = []
labels = []
saved_model.eval()
#regexes - Alīna Govoruhina(2022)
newLine ="\\n|\\r"
urls = '(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})'
numbers = '\d+((\.|\-)\d+)?'
mentions = '\B\@([\w\-]+)'
hashtag = '#'
whitespaces = '\s+'
leadTrailWhitespace = '^\s+|\s+?$'

#combining regexes into single 
combined_pattern = f'{newLine}|{urls}|{numbers}|{mentions}|{hashtag}'

def preprocess(message):
    return message.replace(combined_pattern, '')

for value in df_year['message']:
    print(i)
    i += 1
    message=preprocess(value)
    tokenized_input = tokenizer.encode_plus(
        message,
        max_length=25,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    input_tensor = tokenized_input['input_ids']
    
    #sometimes it puts it on cpu for some reason
    if input_tensor.device.type != 'cuda':
        input_tensor = input_tensor.to('cuda')
    
    # pass the input tensor through the model
    with torch.no_grad():
        output = saved_model(input_tensor)

    # get the index of the predicted class with the highest probability
    #print(output)
    _, predicted = torch.max(output.logits, dim=1)

    # convert the index to a human-readable label
#    if predicted.item() == 0:
#        sentiment = "negative"
#    elif predicted.item() == 1:
#        sentiment = "neutral"
#    elif predicted.item() == 2:
#        sentiment = "positive"

    labels.append(predicted.item())

df_year['label'] = labels
i=0
```
